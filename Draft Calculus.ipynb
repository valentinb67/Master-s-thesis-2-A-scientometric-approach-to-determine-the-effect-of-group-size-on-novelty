```python
DF_MachineLearning_Lee = pd.read_csv('DF_MachineLearning_Lee')
print(DF_MachineLearning_Lee['type'].unique())
DF_SDG_Lee = pd.read_csv('DF_medical_Lee')
print(DF_SDG_Lee['type'].unique())
DF_medical_Lee = pd.read_csv('DF_medical_Lee')
print(DF_medical_Lee['type'].unique())
DF_quantum_Lee = pd.read_csv('DF_quantum_Lee')
print(DF_quantum_Lee['type'].unique())

print(DF_medical_Lee.describe())
print(DF_medical_Lee.describe())
print(DF_quantum_Lee.describe())
```
